```python
import pandas as pd
import chromadb
from src.settings import settings
```

```python
client = chromadb.HttpClient(
    host=settings.CHROMA_HOST,
    port=settings.CHROMA_PORT,
)
```

```python
client.heartbeat()
```

```python
collection = client.get_collection(name=settings.COLLECTION_NAME)
```

```python
collection.count()
```

```python
data = collection.get(
    limit=100,
    offset=0,
    # include=["embeddings"]
)
```

```python
df = pd.DataFrame({
    "id": data["ids"],
    "document": data["documents"],
    "metadata": data["metadatas"],
    "embedding": data["embeddings"],  # список float внутри каждой строки

})
```
